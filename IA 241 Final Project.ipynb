## 2. The Sum of Case Numbers per Day in Bermuda

```python
sum_case_numbers = df.groupby('dateRep').sum()['cases']

sum_case_numbers.plot()
```

## 3. Average Death per Day for One Month in Bermuda

```python
avg_death_per_day = df.groupby('day').mean()['deaths']

avg_death_per_day.plot.bar()
```
